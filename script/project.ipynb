## final Q
- How old were the winners of the Nobel prize?
- Who are the youngest and oldest Nobel Prize winner ?

```python
#  organizathion don't have an age 
final_nobel=nobel[nobel.laureate_type == "Individual"]
final_nobel.info()
```

```python
#  organizathion don't have an age 
final_nobel=nobel[nobel.laureate_type == "Individual"]
f = final_nobel.copy()
f = f.dropna(subset="birth_date")
final_nobel = f
final_nobel.info()
```

```python
#fix the dates 
def fix_date(date_str):
    if date_str.endswith('00-00'):
        return date_str.replace('00-00', '01-01')#add a valide date
    return date_str
#fix the dates
final_nobel.birth_date = final_nobel["birth_date"].convert_dtypes(str).apply(fix_date)
# converting str to time 
final_nobel.birth_date = pd.to_datetime(final_nobel.birth_date)

final_nobel["age_of_winning"] =final_nobel.year - final_nobel.birth_date.dt.year
```

```python
#should use boxplot but this one look better
sns.boxenplot(data=final_nobel,x ="age_of_winning",palette="Spectral")
```

```python
final_nobel.sort_values("age_of_winning",ignore_index=True,inplace=True)
youngest_winner =final_nobel.full_name.iloc[0]
oldest_winner = final_nobel.full_name.iloc[-1]
```
