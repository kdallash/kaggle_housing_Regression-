## converting the object columns to numerical to  be easier to use 

```python
# making a map for str to numbers 
map = {"Ex" :5,	
       "Gd":4,
       "TA":3,
       "Fa":2,
       "Po":1,
       "NA":0,
       "Av":5, 'Mn':3, 
       'No':0, "nan":0,
       'Y':1,'P':0.5,'N':0}
```

```python
# using the map to replace the string with number 
housingdf.replace(map, inplace=True)
```

```python
# Adding the similar category into one  
housingdf["Exter"] = housingdf["ExterQual"]+housingdf["ExterCond"]
housingdf["Bsmt"] = housingdf["BsmtCond"]+housingdf["BsmtQual"] +housingdf["BsmtExposure"]
# dropping unneeded category
housingdf.drop(labels=["ExterQual","ExterCond","BsmtCond","BsmtQual","BsmtExposure"],axis = 1,inplace = True)
```

```python
# Finding the corrilation between the label and data
correlation_metrix =housingdf.corr(numeric_only=True)
correlation_metrix["SalePrice"].sort_values(ascending=False)
# look like extra and basmt have good correlation
```

```python
# making an array with the 20 index which have the highest correlation 
high_corr_index = correlation_metrix["SalePrice"].sort_values(ascending=False).index[0:20]
```

```python
numerical_high_cor_housingdf = housingdf[high_corr_index]
numerical_high_cor_housingdf.info()
```

```python
# drop useless columns 
numerical_high_cor_housingdf.drop(labels="PoolQC",axis=1,inplace=True)     
```

### calculate the skewness of the data 

```python
#plot the data to see if there is a skew  
numerical_high_cor_housingdf.hist(bins=50,figsize=(15,12))
plt.show()
# some data has a right skew like sale price amd TotalBsmtSF
```

```python
y = numerical_high_cor_housingdf[["SalePrice"]]
x = numerical_high_cor_housingdf.drop(labels="SalePrice",axis=1)
```

```python
from sklearn.model_selection import train_test_split
# splitting the data from label
y = numerical_high_cor_housingdf[["SalePrice"]]
X= numerical_high_cor_housingdf.drop(labels="SalePrice",axis=1)


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=24)
```

```python
X_train
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer 
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso,LogisticRegression,LinearRegression,Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PowerTransformer,StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
# Define the pipeline structure
pipeline_template = [
    ("imputer", IterativeImputer()),
    ("scaler", StandardScaler()),
    ("correlator", PowerTransformer(method="yeo-johnson",standardize=False)),
    ("dimensionality_reduction", PCA(n_components=2)),
    ("model", None)  # Placeholder for the model
]
```

```python
# Define models and their parameter grids
models = {
    "Linear Regression": (LinearRegression(), {
        "model__fit_intercept": [True, False],
    }),
    "Ridge Regression": (Ridge(), {
        "model__alpha": [0.1, 1.0, 10.0],
        "model__fit_intercept": [True, False],
    }),
    "Lasso Regression": (Lasso(), {
        "model__alpha": [0.1, 1.0, 10.0],
        "model__fit_intercept": [True, False],
    }),
    "Random Forest": (RandomForestRegressor(), {
        "model__n_estimators": [10, 50, 100],
        "model__max_depth": [None, 10, 20],
    }),
    "DecisionTreeRegressor":(DecisionTreeRegressor(),{
            'model__max_depth': [None, 3, 5,10],
    'model__splitter': ['best', 'random'],
    'model__max_features': ['sqrt', 'log2'],

    })
}
```

```python
# Store results for comparison
results = {}
# Loop through models and perform grid search
for model_name, (model, param_grid) in models.items():
    # Create a new pipeline for each model
    pipeline = Pipeline(steps=pipeline_template)
    pipeline.set_params(model=model)  # Set the current model

    # Perform grid search
    grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring="r2")
    grid_search.fit(X_train, y_train.values.ravel())

    # Store the best scores for all metrics
    results[model_name] = grid_search.best_score_
```

```python
# Print the comparison of results
print("Model Comparison:")
for model_name, scores in results.items():
    print(f"{model_name}  r2 score is {scores}")
```

## Fine toning the random forest model 


```python
param_grid = {
    'model__n_estimators': [100, 300, 500],
    'model__max_depth': [None, 10, 20, 30],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4],
}
```

```python
pipeline_forest = Pipeline(steps=pipeline_template)
pipeline_forest.set_params(model=RandomForestRegressor())  # Set the model to the best model
```

```python
# Perform grid search
grid_search = GridSearchCV(pipeline_forest, param_grid, cv=5, scoring="r2")
grid_search.fit(X_train, y_train.values.ravel())
```

```python
print(grid_search.best_score_)
best_forest = grid_search.best_estimator_
```

```python
# importing the metrics 
from sklearn.metrics import r2_score,mean_absolute_error,max_error
```

```python
def scoring(y_test,predect):
    r2 = r2_score(y_true=y_test,y_pred=predect)
    mean = mean_absolute_error(y_true=y_test,y_pred=predect)
    max = max_error(y_true=y_test,y_pred=predect)
    return(f"r2 score =  {r2},mean absolute error = {mean},max error = {mean}")
```

```python
predict = best_forest.predict(X_test)
print(scoring(y_test=y_test,predect=predict))
```
